```python
import gym
import numpy as np
import yfinance as yf
import pandas as pd
from gym import spaces
from stable_baselines3 import DQN
import matplotlib.pyplot as plt

# ① 株価データ取得（トヨタ）
df = yf.download("7203.T", start="2019-01-01", end="2024-01-01")[["Close"]]
df["MA5"] = df["Close"].rolling(window=5).mean()
df["MA25"] = df["Close"].rolling(window=25).mean()
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

# ② 自作強化学習環境（OpenAI Gym形式）
class StockEnv(gym.Env):
    def __init__(self, df):
        super(StockEnv, self).__init__()
        self.df = df
        self.action_space = spaces.Discrete(3)  # 0: Hold, 1: Buy, 2: Sell
        self.observation_space = spaces.Box(low=0, high=np.inf, shape=(3,), dtype=np.float32)
        self.reset()

    def reset(self):
        self.current_step = 0
        self.stock_held = 0
        self.cash = 1_000_000
        self.total_value = self.cash
        return self._get_obs()

    def _get_obs(self):
        row = self.df.iloc[self.current_step]
        return np.array([row["Close"], row["MA5"], row["MA25"]], dtype=np.float32)

    def step(self, action):
        row = self.df.iloc[self.current_step]
        price = row["Close"]
        reward = 0

        if action == 1 and self.stock_held == 0:  # Buy
            self.stock_held = self.cash // price
            self.cash -= self.stock_held * price
        elif action == 2 and self.stock_held > 0:  # Sell
            self.cash += self.stock_held * price
            reward = self.cash + self.stock_held * price - self.total_value
            self.stock_held = 0

        self.total_value = self.cash + self.stock_held * price
        self.current_step += 1
        done = self.current_step >= len(self.df) - 1
        return self._get_obs(), reward, done, {}

# ③ 環境作成と学習
env = StockEnv(df)
model = DQN("MlpPolicy", env, verbose=0)
model.learn(total_timesteps=10000)

# ④ 学習後のシミュレーション
obs = env.reset()
portfolio_values = []
for _ in range(len(df) - 1):
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, _ = env.step(action)
    portfolio_values.append(env.total_value)
    if done:
        break

# ⑤ 結果の描画
plt.plot(portfolio_values)
plt.title("DQN Portfolio Value (Toyota)")
plt.xlabel("Days")
plt.ylabel("Portfolio Value (JPY)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd, requests, datetime as dt
from pandas_datareader.data import DataReader
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,4)

# ---------- 1. Apple Cash from SEC ----------
CIK = "0000320193"
url  = f"https://data.sec.gov/api/xbrl/companyfacts/CIK{CIK}.json"
hdrs = {"User-Agent": "apple-cash-study/0.1 (your.email@example.com)"}

facts = requests.get(url, headers=hdrs).json()["facts"]["us-gaap"]
cash_items = facts["CashAndCashEquivalentsAtCarryingValue"]["units"]["USD"]

records = []
for item in cash_items:
    if item["form"] in {"10-Q", "10-K"}:          # 四半期 & 年次
        end = pd.to_datetime(item["end"])
        if end >= pd.Timestamp("2019-01-01"):
            records.append({"date": end,
                            "apple_cash_bil_usd": item["val"]/1_000})  # →Billion

apple_cash = (pd.DataFrame(records)
                .drop_duplicates(subset="date", keep="last")
                .set_index("date")
                .sort_index())

# ---------- 2. US M2 from FRED ----------
m2_w = DataReader("WM2NS", "fred",
                  start="2019-01-01", end=dt.date.today())
m2_q = (m2_w.resample("QE").last() / 1_000)       # →Billion
m2_q.rename("m2_bil_usd", inplace=True)           #   (weekly ⇒ 四半期末):contentReference[oaicite:1]{index=1}

# ---------- 3. Align & ratio ----------
df = pd.concat([apple_cash, m2_q], axis=1)
df["m2_bil_usd"] = df["m2_bil_usd"].ffill()       # 前方補完
df["cash_to_m2_ratio"] = df["apple_cash_bil_usd"] / df["m2_bil_usd"]

# ---------- 4. Plot ----------
fig, axs = plt.subplots(3, 1, sharex=True, figsize=(11,9))
df["apple_cash_bil_usd"].plot(ax=axs[0], marker="o",
                              title="Apple Cash and Cash Equivalents (Bn USD)")
axs[0].set_ylabel("Bn USD")

df["m2_bil_usd"].plot(ax=axs[1], marker="o",
                      title="US M2 Money Stock (Bn USD)")
axs[1].set_ylabel("Bn USD")

df["cash_to_m2_ratio"].plot(ax=axs[2], marker="o",
                            title="Apple Cash ÷ US M2")
axs[2].set_ylabel("Ratio")
axs[2].set_xlabel("Quarter‐End")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd, requests, datetime as dt
from pandas_datareader.data import DataReader
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (11, 9)

# ---------- 1. Apple Cash -----------
CIK = "0000320193"
url  = f"https://data.sec.gov/api/xbrl/companyfacts/CIK{CIK}.json"
hdrs = {"User-Agent": "apple-cash-study/0.1 (your.email@example.com)"}

facts = requests.get(url, headers=hdrs).json()["facts"]["us-gaap"]
cash_tags = ["CashAndCashEquivalentsAtCarryingValue",
             "CashAndCashEquivalentsAndShortTermInvestments"]  # 予備タグ

records = []
for tag in cash_tags:
    if tag not in facts:
        continue
    for item in facts[tag]["units"]["USD"]:
        if item["form"] in {"10-Q", "10-K"}:
            end = pd.to_datetime(item["end"])
            if end >= pd.Timestamp("2019-01-01"):
                records.append({"date": end,
                                "apple_cash_bil_usd": item["val"] / 1_000})

apple_cash = (pd.DataFrame(records)
                .drop_duplicates("date", keep="last")
                .set_index("date")
                .sort_index())

# ---------- 2. US M2 ---------------
m2_w = (DataReader("WM2NS", "fred",
                   start="2019-01-01",
                   end=dt.date.today())
          .squeeze())

m2_q = (m2_w
          .resample("QE").last()
          .div(1_000)
          .rename("m2_bil_usd"))

# ---------- 3. 合体 & 比率 ----------
df = pd.concat([apple_cash, m2_q], axis=1)
df["m2_bil_usd"] = df["m2_bil_usd"].ffill()  # M2 最新値で前方補完
df["cash_to_m2_ratio"] = df["apple_cash_bil_usd"] / df["m2_bil_usd"]

# ---------- 4. 可視化 ---------------
fig, axs = plt.subplots(3, 1, sharex=True)

df["apple_cash_bil_usd"].plot(ax=axs[0], marker="o",
                              title="Apple Cash & Cash Equivalents (Bn USD)")
axs[0].set_ylabel("Bn USD")

df["m2_bil_usd"].plot(ax=axs[1], marker="o",
                      title="US M2 Money Stock (Bn USD)")
axs[1].set_ylabel("Bn USD")

df["cash_to_m2_ratio"].plot(ax=axs[2], marker="o",
                            title="Apple Cash ÷ US M2")
axs[2].set_ylabel("Ratio")
axs[2].set_xlabel("Quarter-End")

plt.tight_layout()
plt.show()
```
